###### 마우스를 가져다 대면 활성화되는 탭을 클릭

```python
from selenium import webdriver
import time
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.action_chains import ActionChains
import pandas as pd

driver = webdriver.Chrome(ChromeDriverManager().install())

# url연결해서 창 띄우기
driver.get("https://www.hsd.co.kr/") 

# 창 최대화 하기
driver.maximize_window()
```

```python


# 액션 체인 생성
actions = ActionChains(driver)

# 메뉴탭 xpath가져오기
menu = driver.find_element_by_xpath('//*[@id="gnb"]/div[2]/ul/li[3]/p/a')
all_menu = driver.find_element_by_xpath('//*[@id="gnb"]/div[2]/ul/li[3]/div/ul/li[1]/a')

actions.move_to_element(menu)
actions.click(all_menu)
time.sleep(2)

# 액션실행
actions.perform()

```

```python
# 더보기 버튼 생성
more = driver.find_element_by_css_selector('#btn_more > span > a')

# while을 이용한 더보기 루프 생성
while True:
    try:
        # 더보기 버튼 클릭하기
        more.click()
        # 너무 빠르면 중복이 일어날 수 있음. 너무 오래 설정하면 크롤링 속도가 느려짐.
        time.sleep(0.1)
    except:
        break
```

```python
title_list = []
price_list = []

# 메뉴 가져오기
title = driver.find_elements_by_css_selector('h4.h.fz_03')
for i in title:
    title_list.append(i.text)
    # print(i.text)

print("")
# 가격 가져오기
#menuList_456 > div > div.item-text > div > strong
price = driver.find_elements_by_css_selector('div.item-price > strong')
for i in price:
    price_list.append(i.text)
#     print(i.text)

data = {"메뉴명" : title_list, "가격" : price_list}

food = pd.DataFrame(data)
food


```
