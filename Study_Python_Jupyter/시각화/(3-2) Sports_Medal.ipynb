```python
import pandas  as pd
import numpy as np
```

```python
df = pd.read_csv("winter.csv", encoding = "cp949")
df
```

```python
#데이터에서 한국(KOR) 데이터만 추출
kr = df[df.Country == "KOR"]
kr.head(3)
```

<u>피벗테이블</u> 만들기

```python
a = kr.pivot_table(index = "Year", columns="Medal", aggfunc="size")
# 연도별로 메달 개수 알고싶음
# 연도별 인덱스 설정 후, 알고 싶은 정보의 컬럼은 columns으로, aggfunc에 원하는 함수
a
```

```python
a
```

```python
# 결측치 값을 0으로 채우기
# a.fillna()
a.loc[a.Bronze.isnull(), "Bronze"] = 0
```

```python
#피벗테이블 : 스포츠별로 성별 수를 구해보기
s = df.pivot_table(index="Sport", columns="Gender", aggfunc="size")
s.head(3)
```

```python
#discipline별로 메달수 구하기
d = df.pivot_table(index="Discipline", columns="Medal", aggfunc="size")
d.head(4)
```
