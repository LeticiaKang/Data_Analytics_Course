### 시각화 Matplotlib


```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

#### Line Plot 그리기

```python
x = [1,2,3]
y = [2,4,6]
plt.plot(x, y)
plt.show()
```

#### Line Style 설정하기

![선의 종류](https://user-images.githubusercontent.com/87592790/191677097-b9910e68-434d-4b65-bee0-188fc32d1f5c.png)



```python
x = [1,2,3]
y = [2,4,6]
plt.plot(x, y, ls = '--')
plt.show()
```

#### Marker찍기
<a>https://matplotlib.org/stable/api/markers_api.html</a>

![선의 종류](https://user-images.githubusercontent.com/87592790/191677524-9cac7f64-f7e8-44dd-8a9b-f71a691c2c58.png)


```python
x = [1,2,3]
y = [2,4,6]
plt.plot(x, y, ls = '--', marker = "o")
plt.show()
```
