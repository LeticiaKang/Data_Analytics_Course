#### Line width : 선의 굵기 / lw

```python
x = [1,2,3]
y = [2,4,6]
plt.plot(x, y, ls = '--', marker = "o", lw = 5)
plt.show()
```

#### Line Color : 선의 색 / color

![선의 종류](https://user-images.githubusercontent.com/87592790/191680667-076406b6-f97b-45e2-97ad-97d0b5706a2f.png)

```python
x = [1,2,3]
y = [2,4,6]
plt.plot(x, y, ls = '--', marker = "o", lw = 5, color = "red")
plt.show()
```

```python
x = [1,2,3]
y = [2,4,6]
plt.plot(x, y, ls = '--', marker = "o", lw = 5
         , color = "black", markerfacecolor = "r")
plt.show()
```

#### 이 외의 스타일 옵션

![선의 종류](https://user-images.githubusercontent.com/87592790/191680843-2755ee54-58bd-47b1-834c-0dc34b6b7341.png)


```python
x = [1,2,3]
y = [2,4,6]
plt.plot(x, y, ls = '--', lw = 5, color = "black"
         , marker = "o", mec = "r"
         , ms = 50, mew = 100, mfc = "y"
         )
plt.show()
```

### 그림 범위 지정
- plt.xlim(시작값, 끝값)

```python
x = np.arange(8)
y = [5,2,3,6,8,7,5,7]

plt.plot(x,y, ls = "-.", c = "m", marker = "<", mfc = "k" , mec = "k", ms = 10)

plt.xlim(-2,10)
plt.ylim(-5,10)

plt.show()
```

### 그리드 그리기 : plt.grid()

```python
x = np.arange(8)
y = [5,2,3,6,8,7,5,7]

plt.plot(x,y
         , ls = "-."
         , c = "m"
         , marker = "<"
         , mfc = "k" 
         , mec = "k"
         , ms = 10)

plt.xlim(-2,10)
plt.ylim(-5,10)

plt.grid()

plt.show()
```

![선의 종류](https://user-images.githubusercontent.com/87592790/191684036-91354fc2-90ad-4f62-94b3-ec3857022553.png)



```python
x = [0,1,2,3,4,5,6,7]
y = [5,2,3,6,8,7,5,7]

plt.plot(x,y, ls = "--", lw = 5, marker = "o"
        , ms = 15, mec = "g", mew = 5, mfc = "r")
plt.xticks([0,3,5])
plt.yticks([1,5,9])
plt.show()
```

#### * 제목달기 : plt.title("제목입니다")
#### * X, Y축 title 표시 : plt.xlabel("Y축").set_rotation(0)

```python
#### 범례표시 / 한글 적용
from matplotlib import  font_manager, rc
font_name = font_manager.FontProperties(fname = "C:/Windows/Fonts/malgun.ttf").get_name()
# matplotlib.font_manager("font", family = font_name)  #적용이 안됨
rc("font", family = font_name)

x = [1,2,3,4]
y = [2,4,6,8]
z = [3,6,9,12]

plt.plot(x, label = "엑스")
plt.plot(y, label = "와이")
plt.plot(z, label = "제트")

plt.xlabel("X축")
plt.ylabel("Y축").set_rotation(2880)

plt.title("제목입니다")

plt.legend() #범례를 화면에 표시해주는 

plt.show()
```
