## Question2

```python
X = [2,3,4,5,6,7,8]
Y1 = [5,7,3,9,1,8,6]
Y2 = [9,5,7,2,3,7,1]
```

```python
font_name = font_manager.FontProperties(fname = "C:/Windows/Fonts/malgun.ttf").get_name()
rc("font", family = font_name)

#그래프 내부 크키
plt.figure(figsize = (5,5)).patch.set_facecolor("w") #배경색

plt.plot(    X, Y1
            , label = "Y1"
            , c = 'r'
        )

plt.plot(   X, Y2
            , label = "Y2"
            , c = 'b'
        )

plt.legend(loc='best')

#축 레이블(이름) 설정하기 : plt.x(y)label("이름")
plt.xlabel("X"       #x축 이름
           , loc='center'       #레이블 위치 설정하기 => {‘left’, ‘center’, ‘right’} 中 택1
           )   
plt.ylabel("Y"     #y축 이름
           , loc = "center"       #레이블 위치 설정하기 => {‘bottom’, ‘center’, ‘top’}
           )      

#그래프 이름 설정하기
plt.title("X와 Y1,Y2의 관계")

plt.grid()

plt.show()
```
