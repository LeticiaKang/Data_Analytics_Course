```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.action_chains import ActionChains
import pandas as pd #엑셀화 판다스
from selenium.webdriver.common.keys import Keys #키보드
from bs4 import BeautifulSoup as bs
```

```python
driver = webdriver.Chrome(ChromeDriverManager().install())
driver.get("https://www.youtube.com/c/%EB%AF%B8%EB%AF%B8%EB%AF%B8%EB%88%84/videos") 

# 창 최대화 하기
driver.maximize_window()

title_list = []
view_num_list = []
category_list = []

# body tag는 1개라서 element로 넣어줘야 함
body = driver.find_element_by_css_selector("body")

while True:
    # last : 스크롤 전 객체화된 부분
    last = bs(driver.page_source, "lxml")
    # 10번 스크롤 실행
    for i in range(10):
        # 페이지를 내리겠다는 send_keys
        body.send_keys(Keys.END)    
    
    # new : 스크롤링 된 후 객체화된 부분
    new = bs(driver.page_source, "lxml")
    if last == new:
        break

title = driver.find_elements_by_css_selector("#video-title")
view_num = driver.find_elements_by_css_selector("#metadata-line > span:nth-child(1)")

```

```python

for i in title:
    title_list.append(i.text)
    
for i in view_num:
    view_num_list.append(i.text.replace("조회수 ", ""))
```

```python
len(title_list), len(view_num_list)
```

```python
num = list(range(1,len(title_list)+1))

data = {"인덱스" : num, "제목" : title_list, "조회수" : view_num_list}
mimiminu = pd.DataFrame(data)
mimiminu.set_index("인덱스")
```
