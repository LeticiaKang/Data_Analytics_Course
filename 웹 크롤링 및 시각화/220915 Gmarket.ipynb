```python
from selenium import webdriver
import time
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.action_chains import ActionChains
import pandas as pd
```

```python
driver = webdriver.Chrome(ChromeDriverManager().install())

# url연결해서 창 띄우기
driver.get("https://corners.gmarket.co.kr/Bestsellers") 

# 창 최대화 하기
driver.maximize_window()

# 액션 체인 생성
actions = ActionChains(driver)

title_list = []
price_list = []
category_list = []


for i in range(200):
    try:
        # 아이템 페이지 마다 i가 달라짐
        item = driver.find_elements_by_css_selector('img.lazy') # 이미지
        item[i].click() # 해당 css 클릭
        driver.implicitly_wait(5)
        # 이동한 페이지에서 원하는 정보 가져오는 seletor 선언하기
        category = driver.find_element_by_css_selector('li.on > a')
        title = driver.find_element_by_css_selector('h1.itemtit')
        price = driver.find_element_by_css_selector('strong.price_real')
        # 원하는 정보 리스트에 넣기
        category_list.append(category.text)
        price_list.append(price.text.replace("원", ""))
        title_list.append(title.text)
        
        driver.back()
        driver.implicitly_wait(5)
    except:
        print(f"{i}번째 오류~!")
        pass
```

```python
num = list(range(1,201))

data = {"순위" : num, "카테고리명" : category_list, "상품명" : title_list, "가격" : price_list}

best_items = pd.DataFrame(data)
best_items.set_index("순위")
```
