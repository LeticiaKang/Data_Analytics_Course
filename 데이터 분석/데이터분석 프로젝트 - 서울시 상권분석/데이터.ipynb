### 학원
##### - 시장규모와 업체수가 상관성 높음(=밀집도 높음)
##### - 시장규모가 클수록 영업이익을 높음
##### - 학원 창업시 시장규모가 큰(밀집도가 높음) 위치로 들어가는 것이 좋음
##### - 유동인구는 관계 없음

```python
# 상관관계 확인
academic_re.corr()
academic_re.describe()

# 매출 상위 2% 업체 (N+2∂)
top_2per = academic_re[academic_re["창업 후 평균 영업이익"] >= (academic_re.describe().iloc[1,0] + 2*academic_re.describe().iloc[2,0])]
top_2per.sort_values(by="창업 후 평균 영업이익", ascending=False)
```
