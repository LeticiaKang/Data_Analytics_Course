문제1. 00중학교 1-1반의 중간고서 평균 성적(국,영,수,과,사)점수를
1. 시리즈로 생성해주세요
2. 국어 75, 영어 80, 수학 90, 과학 95, 사회 70 
3. 시리즈 이름은 "중간고사 성적"
4. 인덱스 이름은 과목, 인덱스는 국어, 영어, 수학, 과학, 사회로 설정

```python
midTerm_score = pd.Series([75, 80, 90, 95, 70], 
                    index=["국어", "영어", "수학", "과학", "사회"],
                    # columns = ["과목", ""],
                    name = "중간고사 성적")
midTerm_score.index.name = "과목"
# midTerm_score.name = "중간고사 성적"  #시리즈 이름변경 방법2
midTerm_score
```

문제2. 국어, 영어, 수학 행만 보여주세요

```python
#방법1 : 인덱스 이름 이용
midTerm_score["국어" : "수학"]
#방법2 : 인덱스 번호 이용
midTerm_score[0 : 3]
#방법3 : 인덱스 이름 전부 나열
midTerm_score[["국어", "영어", "수학"]]
```

문제3. 중간고사 평균 성적이 85이상인 과목만 추출해주세요

```python
midTerm_score[midTerm_score >= 85]
#midTerm_score[조건]을 넣으면 boolean형으로 나온다는 것을 기억하기!
```

문제4. 평균성적이 80점 이상 90점이하인 과목만 추출

```python
midTerm_score[(midTerm_score >= 80) & (midTerm_score <= 90)]
```

문제5. 5개의 과목 평균, 합계, 최대값을 출력해주세요

```python
print(f"평균 점수 : {midTerm_score.mean()}")
print(f"합계 점수 : {midTerm_score.sum()}")
print(f"최고 점수 : {midTerm_score.max()}")
print(f"최고 점수의 과목명 : {midTerm_score[midTerm_score == midTerm_score.max()].index[0]}")
```
