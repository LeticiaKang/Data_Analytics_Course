1.결측치 처리하는 방법 : __*제거*__ 하거나 __*다른 값으로 채우기*__

```python
import pandas as pd
import numpy as np
```

```python
d = {'score1':[100, 90, np.nan, 95],
     'score2':[30, np.nan, 45, 56],
     'score3':[52, 40, 80, 98],
     'score4':[np.nan, np.nan, np.nan, 65]
     }

df = pd.DataFrame(d)
df
```

```python
# df.info()
# 각각 컬럼별 null값의 개수를 알려줌
df.isnull().sum()
```

```python
# null이 있는 행 출력
df[df['score1'].isnull()]

# 행을 기준으로 삭제하기
df.dropna()

# how='any' : 결측치가 있는 행 전부를 지운다
df.dropna(how='any')
```

```python
# 열을 기준으로 삭제하기
df.dropna(axis=1)

# 행의 전체값이 결측치인 행을 삭제하기
d2 = {'score1' : [100, np.nan, np.nan, 95],
      'score2' : [np.nan, np.nan, np.nan, np.nan],
      'score3' : [52, np.nan, 80, 98],
      'score4' : [np.nan, np.nan, np.nan, 65]
      }

# 데이터 프레임 만들어서 결측치 다루기
df2 = pd.DataFrame(d2)

# 모든 행의 값이 결측치일 경우만 지우고 싶을 때,
df2.dropna(how='all')

# 임계치를 사용해서 몇 개이상 있을 경우 제거하기
# ** 적당한 임계치가 중요함! 너무 많이도 적지도 않게 사용하려면
df2.dropna(thresh=2)

# 특정 열 안에서만 삭제하고 싶을 때
df2.dropna(subset=['score2', 'score3'])
```

```python
# 결측치 채우기
df.fillna(0)

# 결측치 바로 이전의 값으로 채우기
# ※하지만! 앞에 값이 NaN이면 채워지지 않는다!!
df.fillna(method='pad')

# 결측치 바로 뒤의 값으로 채우기!
df.fillna(method='bfill')

# 결측치를 각 열의 평균값으로 채우기
df.fillna(df.mean())
```

채우기!

```python
# 앞과 뒤의 중간값으로 채워줌
df.interpolate(method='linear', limit_direction = 'forward')

# df
df.interpolate(method='pad', limit=2)
```
