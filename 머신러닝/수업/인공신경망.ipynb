```python
import mglearn
import graphviz
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import make_moons
```

# 학습 목표
### 1.인공신경망에는 어떤 하이퍼파라미터가 있는지 알 수 있다.
### 2.인공신경망의 결정경계를 확인하고 
### 3.결정경계를 부드럽게 하기 위해서 하이퍼파라미터를 사용하여 조절할 수 있다.

- 인공신경망에 tew_moons 데이터셋을 적용
- 은닉 유닛의 개수는 따로 X로 설정하고, 기본값 100사용한다.

```python
# 데이터 불러오기
x, y = make_moons(n_samples=100, noise=0.25, random_state=0) #데이터 100개만 사용할거야

# 데이터 분리하기
X_train, X_test, y_train, y_test = train_test_split(x, y, 
                                                    stratify = y,
                                                    random_state=42)

# 모델 불러오기(분류모델, MLP)
mip = MLPClassifier(random_state=0)

# 모델 학습
mip.fit(X_train, y_train)

# 모델이 그래프로 결정경계 확인하기
mglearn.plots.plot_2d_separator(mip, X_train, fill=True, alpha=.3)
mglearn.discrete_scatter(X_train[:, 0], X_train[:, 1], y_train)
plt.xlabel("property 0")
plt.ylabel("property 1")
```

```python
# hidden_layer_sizes = [10] #층이 여러가면 ()튜플에 작성

# 모델 불러오기(분류모델, MLP)
mip = MLPClassifier(random_state=0, hidden_layer_sizes=10)

# 모델 학습
mip.fit(X_train, y_train)

# 모델이 그래프로 결정경계 확인하기
mglearn.plots.plot_2d_separator(mip, X_train, fill=True, alpha=.3)
mglearn.discrete_scatter(X_train[:, 0], X_train[:, 1], y_train)
plt.xlabel("property 0")
plt.ylabel("property 1")
```

```python
# 결정경계가 부드럽지 않기 때문에,
# 은닉유닛이나 은닉층을 추가하거나 활성화 함수를 사용한다.

# < 10개의 유닛으로 구성된 두 개의 은닉층 >

# 모델 불러오기(분류모델, MLP)
mip = MLPClassifier(random_state=0, hidden_layer_sizes=[10, 10])

# 모델 학습
mip.fit(X_train, y_train)

# 모델이 그래프로 결정경계 확인하기
mglearn.plots.plot_2d_separator(mip, X_train, fill=True, alpha=.3)
mglearn.discrete_scatter(X_train[:, 0], X_train[:, 1], y_train)
plt.xlabel("property 0")
plt.ylabel("property 1")
```

```python
# < 10개의 유닛으로 구성된 두 개의 은닉층 & 활성화 함수 적용하기 >

# 모델 불러오기(분류모델, MLP)
mip = MLPClassifier(random_state=0, hidden_layer_sizes=[10, 10]
                    , activation = 'tanh')

# 모델 학습
mip.fit(X_train, y_train)

# 모델이 그래프로 결정경계 확인하기
mglearn.plots.plot_2d_separator(mip, X_train, fill=True, alpha=.3)
mglearn.discrete_scatter(X_train[:, 0], X_train[:, 1], y_train)
plt.xlabel("property 0")
plt.ylabel("property 1")
```
