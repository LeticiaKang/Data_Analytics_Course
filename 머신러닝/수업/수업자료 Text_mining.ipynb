```python
import nltk
from nltk.corpus import stopwords
nltk.download('book', quiet= True)
from nltk.book import *
from nltk.tokenize import RegexpTokenizer, WordPunctTokenizer
from collections import Counter
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
# 목록 확인
nltk.corpus.gutenberg.fileids()
```

```python
data = nltk.corpus.gutenberg.raw('austen-emma.txt')
data = data[:1000]
data
```

```python
# Tokenizer (단어 단위로 토크나이징)
tokenizer = WordPunctTokenizer() 
data = tokenizer.tokenize(data) #띄어쓰기 단위로 토크나이징
data
```

```python
# 불용어 제거
data = [each_word for each_word in data if each_word not in stopwords.words()]
```

```python
# 단어 별 언급 빈도 확인
data_cnt = Counter(data)
data_cnt = dict(data_cnt)
data_cnt
```

```python
# Graph
data_graph = nltk.Text(data, name = 'word count')
data_graph.plot()
```

```python
# Word Cloud
WordCloud = WordCloud().generate_from_frequencies(data_cnt)
plt.figure(figsize = (8,4))
plt.imshow(WordCloud)
plt.axis('on')

plt.show()
```

    단어 리스트 생성
    

```python
word_list = df['document'][:3000]
```

```python
tag = []

for s in word_list:
    morph = okt.pos(s)
    tag.append(morph)
```

```python
# 명사 추출
noun_list = []
for s in tag :
    for word, tag in s:
        if tag in ['Noun']:
            noun_list.append(word)
            
            
        
    
```
